```python
import evaluations
import numpy as np
```

```python

testing_x_location = "FYP/testing_x.npy"
testing_y_location = "FYP/testing_y.npy"

VGG_Model_Location = 'FYP/models/VGG16_Model'
VGG_Weights_Location = "FYP/models/VGG_BEST"
VGG_training_history = "FYP/histories/VGG16_history_30Epochs.npy"


Custom_Model_Location = 'FYP/models/Custom_Model'
Custom_Training_History = "FYP/histories/Custom_50Epochs.npy"


ResNet_Model_Location = 'FYP/models/ResNet50_Model'
ResNet_History = "FYP/histories/ResNet_50Epochs.npy"
```

```python
testing_x, testing_y = evaluations.loadTestingData(testing_x_location, testing_y_location)
```

# VGG Eval 

```python
#VGG Predictions
vgg_predictions = evaluations.generatePredictions(VGG_Model_Location, testing_x, VGG_Weights_Location)
```

```python
vgg_class_report_Multi = evaluations.classificationReport(testing_y, vgg_predictions)
print(vgg_class_report_Multi)
```

```python
vgg_class_report_Binary = evaluations.classificationReport(testing_y, vgg_predictions, problem_type = "Binary")
print(vgg_class_report_Binary)
vgg_class_report_Binary_score = evaluations.calculateBinaryScore(vgg_predictions, testing_y)
```

```python
vgg_confusion_matrix_multi = evaluations.plotConfusionMatrix(testing_y, vgg_predictions)
```

```python
vgg_confusion_matrix_binary = evaluations.plotConfusionMatrix(testing_y, vgg_predictions, problem_type = "Binary")
```
